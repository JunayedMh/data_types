A quiz player's scores per round are rounds = (15, 20, 8, 18, 25, 12). Find their total score and average score per round.

Given scores = [10, 20, 30, 40, 50], calculate a weighted total where each score is multiplied by its
position index (0, 1, 2, ...). Use enumerate().

```python
scores = [10, 20, 30, 40, 50]
weighest_total = 0

for i, score in enumerate(scores):
    weighest_total += score* i

print("weighted total: ", weighest_total)
```

Given rounds = [("Round 1", 15), ("Round 2", 20), ("Round 3", 8), ("Round 4", 18)],
find the round where the player scored the least.


```python
rounds = [("Round 1", 15), ("Round 2", 20), ("Round 3", 8), ("Round 4", 18)]

least_round = rounds[0][0]
least_score = rounds[0][1]

for round, score in rounds:
    if score < least_score:
        least_score = score
        least_round = round

print(f"{least_round} has the lowest score: {least_score}")
```

Given results = [("Ali", 80), ("Sara", 95), ("Rahim", 74), ("Mia", 88)], build a dict
mapping each student to their rank (1 = highest). Tip: sort the list by score descending first.

```python
results = [("Ali", 80), ("Sara", 95), ("Rahim", 74), ("Mia", 88)]

student = results[0][0]
highest_mark = results[0][1]

for st, mark in results:
    if mark > highest_mark:
        highest_mark = mark
        student = st

print(student)
print(highest_mark)
```

From orders = [("T-shirt", 2), ("Laptop", 15), ("Pen", 1), ("Monitor", 8), ("Book",
3), ("Phone", 12)], build a dict grouping items into "small" (qty 1-3) and "large" (qty 4+).

```python
orders = [("T-shirt", 2), ("Laptop", 15), ("Pen", 1), ("Monitor", 8), ("Book",
3), ("Phone", 12)]

grouped = {"small": [], "large": []}

for item, qty in orders:
    if qty > 4:
        grouped["large"].append(item)
    else:
        grouped["small"].append(item)

print(grouped)
```
