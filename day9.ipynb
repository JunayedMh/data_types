1. Write a function that takes a list of prices and returns the sum. 

```python
def total_price(prices):
    if not prices:
        return None
    total = 0
    for price in prices:
        total += price
    return total

print(total_price([250,454,324]))
print(total_price([]))
```

2. Write a function that returns the total after subtracting a discount percentage. If no discount is given, it should behave like a plain total.

```python
def total_with_discount(prices, discount=0):
    total = 0

    for price in prices:
        total += price
    return total - (total*discount/100)

print(total_with_discount([123,454], discount=10))
print(total_with_discount([843,322]))
```

3. Write a helper function that returns square of a number. Then write sum_of_squares(numbers) that calls square() inside its loop instead of writing x*x directly.

```python
def square(num):
    return num*num

def sum_of_squares(numbers):
    total = 0
    for number in numbers:
        total += square(number)
    return total

print(sum_of_squares([1,2,3,4]))
```

4. Write a function that return a NEW list where each position holds the running total up to that point.

```python
def current_sum_of_num(numbers):
    list_with_index = []
    total = 0

    for index,number in enumerate(numbers):
        total += number
        list_with_index.append(total)
    return list_with_index

print(current_sum_of_num([25,13,22]))
```

5. Write a function that accepts ANY number of seperate arguments (not a list!) and returns their sum.

```python
def total_sum(*args):
    total = 0

    for num in args:
        total += num
    return total

print(total_sum(4,6,32))
```

6. Write a function to find out the max.

```python
def find_max(nums):
    max_num = nums[0]

    for num in nums:
        if num > max_num:
            max_num=num
    return max_num

print(find_max([5,8,9,6,7,3]))
```

7. Write a function to find out the min number with index.

```python
def find_min(numbers):
    min_num = numbers[0]

    for i in range(len(numbers)):
        if numbers[i] < min_num:
            min_num = numbers[i]
            index_num = i
    return (min_num,index_num)

print(find_min([4,3,6,2,6,2,5,22,3]))
```

8. Write a helper function that return average. Then write another function that calles it and counts how many values are above that average.

```python
def calc_avg(numbers):
    return sum(numbers)/len(numbers)

def count_above_avg(numbers):
    avg = calc_avg(numbers)
    count = 0

    for num in numbers:
        if num > avg:
            count+=1
    return count

print(count_above_avg([32,34,21,5,43]))
```
