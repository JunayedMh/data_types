1. Write a function calculate_total(items) that takes a list of numbers and returns their total. Call it with three different lists.

```python
def calculate_total(nums):
    total = 0
    for num in nums:
        total += num
    return total

print(calculate_total([10,20,30]))
```

2. Write sum_dict_values(data) that takes a dict and returns the sum of all its values. Call it with a sales dict and an inventory dict.

```python
def sum_dict_values(data):
    total = 0
    for value in data.values():
        total += value
    return total 

sales = {"Mon": 1500, "Tue": 2300, "Wed": 1800, "Thu": 2700}
inventory = {"rice": 50, "oil": 8, "salt": 200, "flour": 5}

print("Total sales: ",sum_dict_values(sales))
print("Total inventory: ", sum_dict_values(inventory))
```

3. Write a tax function. Default tax is 8% but the caller can pass a different rate.

```python
def total_with_tax(items,tax_rate=0.08):
    subtotal = 0
    for item in items:
        subtotal += item
    return round(subtotal * (1+tax_rate),2)

cart = [500, 900, 1200, 300]
print(total_with_tax(cart)) # default 8%
print(total_with_tax(cart, 0.15)) # tax changed to 15%
print(total_with_tax(cart,0.0)) # No tax
```

4. Write a fucntion to calculate weighted average. Add a guard: if lists are different length, return None.

```python
def weighted_avg(values, weights):
    if len(values) != len(weights):
        print("Warning: lists must be the same length")
        return None
    total = 0
    for value,weight in zip(values, weights):
        total += value * weight 
    return round(total,2)

grades = [85, 72, 90, 78]
weights = [0.3, 0.2, 0.3, 0.2]
print(weighted_avg(grades, weights))
print(weighted_avg([80, 90], [0.5, 0.3]))
print(weighted_avg([90,89], [0.3,0.2,0.1]))
```

5. Write a function that return a new list where each element is the running total up to that point.

```python
def coumulative_sum(values):
    sum_list = []
    running_sum = 0

    for value in values:
        running_sum += value
        sum_list.append(running_sum)
    return sum_list

daily_sales = [200, 150, 300, 250, 180]
weekly_hits = [1200, 980, 1450, 870, 2100]

print(coumulative_sum(daily_sales))
print(coumulative_sum(weekly_hits))
```

6. Write a function that returns the largest value without using max(). Return None if list empty.

```python
def is_largest(nums):
    if not nums:
        return None

    largest = nums[0]

    for num in nums:
        if num > largest:
            largest = num
    return largest

print(is_largest([15, 42, 8, 73, 29])) 
print(is_largest([]))
print(is_largest([-5, -1, -8, -2]))
```

7. Write a function that takes dict and return the key with the highest value.

```python
def highest_values(values):
    if not values:
        return None
    highest_value = 0

    for key,value in values.items():
        if value > highest_value:
            highest_value = value
            the_key = key
    return the_key

sales = {"Mon": 1500, "Tue": 2300, "Wed": 1800, "Thu": 2700}
scores = {"Ali": 80, "Sara": 95, "Rahim": 74}

print("Best sales day:", highest_values(sales))
print("Top student: ", highest_values(scores))
```

8. Write a function return both the maximum value and its index as a tuple.

```python
def find_max_with_index(values):
    if not values:
        return None, None
    
    max_val = values[0]
    max_index = 0

    for index, value in enumerate(values):
        if value > max_val:
            max_val = value
            max_index = index
    return max_val,max_index

scores = [55, 82, 91, 67, 78, 95, 60]
value,index = find_max_with_index(scores)
print(f"Max score: {value} at index {index}")
```

9. Write a function that keeps value within a range.

```python
def clamp(value,low,high):
    if value< low:
        return low
    if value>high:
        return high
    return value

print(clamp(5, 0, 10))
print(clamp(-3, 0, 10))
print(clamp(15, 0, 10)) 

raw_scores = [-5, 45, 112, 78, 0, 100, 95, -2]
clamped = [clamp(s,0,100) for s in raw_scores]
print("Clamped: ",clamped)
```

10. Write a function that returns the top N(key, value) pairs sorted by value descending.

```python
def top_n_items(data, n=3):
    sorted_items = sorted(data.items(), key=lambda x:x[1], reverse= True)
    return sorted_items[:n]

sales = {"Mon":1500,"Tue":2300,"Wed":1800,"Thu":2700,"Fri":3100,"Sat":900}

print("Top 3:", top_n_items(sales))
```

11. Write a function that counts how many values are stricly above the threshold =0.

```python
def count_above(values):
    count = 0
    threshold = 0

    for value in values:
        if value > threshold:
            count += 1
    return count

print(count_above([1,3,5,-5,0,4,6,9]))
```

12. Write a function that counts how many times a specific value appears.

```python
def count_matching(input,target):
    count = 0

    for item in input:
        if item == target:
            count += 1
    return count

tags = ["python","data","python","ml","data","ai","ml","python"]
orders = ["completed","cancelled","completed","completed","cancelled"]

print(count_matching(tags, "python"))
print(count_matching(orders, "completed"))
```

13. write a function that return True if any value appears more than once.

```python
def has_duplicate(items):
    seen = set()

    for item in items:
        if item in seen:
            return True
        seen.add(item)
    return False

print(has_duplicate([1, 2, 3, 4, 5]))
print(has_duplicate([1, 2, 3, 2, 5]))
print(has_duplicate(["ali","sara","ali"]))
```
