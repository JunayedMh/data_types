```python
# You have prices = (120, 350, 89, 450, 75). Write a loop to find the total cost.

prices = (120, 350, 89, 450, 75)
total = 0

for price in prices:
    total += price

print("total cost = ", total)
```

```python
# You have stock = {"apples": 40, "bananas": 25, "mangoes": 60}. Find the total number of fruits in stock.

stock = {"apples": 40, "bananas": 25, "mangoes": 60}
total_fruits = 0

for quantity in stock.values():
    total_fruits += quantity

print("Total number of fruits: ", total_fruits)
```
