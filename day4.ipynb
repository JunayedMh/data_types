1. Given a list of sentences: lines = ["Python is fun", "I love coding", "Data science is
powerful"]. Count the total number of words across all sentences.


```python
lines = ["Python is fun", "I love coding", "Data science is powerful"]

total_words = 0

for line in lines:
    words = line.split()
    total_words += len(words)

print("Total Words: ",total_words)
```

2. You have two lists: items = ["rice", "oil", "sugar"] and prices = [80, 150, 60]. Use zip() to
calculate the total cost.

*zip()* function in Python like a zipper on a jacket. It takes two or more separate lists (or tuples) and pairs up their corresponding elements together, side-by-side.


```python
items = ["rice", "oil", "sugar"]
prices = [80, 150, 60]
total = 0

for item,price in zip(items,prices):
    total += price
    print(f"{item}: {price}")

print("Total cost: ",total)
```

```python
item_price = {}
total_p = 0
for i in range(len(items)):
    item_price[items[i]] = prices[i]
print(item_price)

for item,price in item_price.items():
    total_p += price

print("Total price: ",total_p)
```

4. Given cart = {"shirt": 500, "pants": 900, "shoes": 1200, "cap": 300} and a 15% discount
on everything, calculate total amount saved.

```python
cart = {"shirt": 500, "pants": 900, "shoes": 1200, "cap": 300}
total = 0

for item,price in cart.items():
    total += price
    discount = total * (15/100)

print("Total saved:",discount)
```

5. Scores are stored as rounds = [("R1", 20, False), ("R2", 35, True), ("R3", 15, False),
("R4", 40, True)] where the third element means it is a bonus round (score counts double). Calculate the
true total.

```python
rounds = [("R1", 20, False), ("R2", 35, True), ("R3", 15, False), ("R4", 40, True)]
total = 0

for name, score, is_bonus in rounds:
    if is_bonus:
        total += score *2
    else:
        total += score

print("Total score: ",total)
```

7. iven items = ["rice", "oil", "sugar", "flour", "salt"] and prices = [80, 150, 60,
55, 20], find the most expensive item using zip().

```python
items = ["rice", "oil", "sugar", "flour", "salt"]
prices = [80, 150, 60, 55, 20]
expensive_item = 0

for item,price in zip(items,prices):
    if price > expensive_item:
        expensive_item = price
        expensive_p = item

print(f"{expensive_p} is the expensive product: {expensive_item}")
```

9. Given words = ("cat", "elephant", "tiger", "ox", "horse", "crow"), find the word whose
length is closest to 5 characters

```python
words = ("cat", "elephant", "tiger", "ox", "horse" "crow")
c_word = words[0]

for word in words:
    if abs(len(word)-5) < abs(len(c_word)-5):
        c_word = word

print(c_word)
```

10. Given grades = {"Math": [78, 85, 90, 72], "English": [88, 76, 92, 84], "Science":
[65, 70, 68, 74]}, find which subject has the highest class average

```python
grades = {"Math": [78, 85, 90, 72], 
          "English": [88, 76, 92, 84], 
          "Science": [65, 70, 68, 74]
          }

top_avg = 0

for subject, marks in grades.items():
    total = 0
    for mark in marks:
        total += mark
    avg = total / len(marks)
    if avg > top_avg:
        top_avg = avg
        top_sub = subject

print(f"{top_sub} got the highest avg: {top_avg}")
```

11. Given words = ["apple", "Banana", "orange", "grape", "avocado", "melon",
"Elderberry"], count how many words start with a vowe

```python
words = ["apple", "Banana", "orange", "grape", "avocado", "melon", "Elderberry"]

vowels = "aeiou"
count = 0

for word in words:
    if word[0].lower() in vowels:
        count += 1

print("Total words start with vowel:", count)
```

15. Given tags = ["python", "data", "python", "ml", "data", "ai", "ml", "python"], count
how many items are duplicates (appear more than once).

```python
tags = ["python", "data", "python", "ml", "data", "ai", "ml", "python"]

TotalTags = len(tags)
UniqueTags = len(set(tags))
DuplicateItems = TotalTags - UniqueTags

print(DuplicateItems)
```

16. Given text = "the cat sat on the mat the cat ate the rat", build a dictionary counting how
many times each word appears.


```python
text = "the cat sat on the mat the cat ate the rat"
words = text.split()
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print(freq)
```

17. Given students = ["Ali", "Sara", "Rahim", "Mia"] and ids = ["S001", "S002", "S003",
"S004"], build a dict mapping student ID to name.

```python
students = ["Ali", "Sara", "Rahim", "Mia"]
ids = ["S001", "S002", "S003","S004"]

lookup = {}

for i,student in enumerate(students):
    lookup[student] = ids[i]

print(lookup)
```
