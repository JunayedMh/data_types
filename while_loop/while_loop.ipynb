2. Start at 100. Keep subtracting 7 each step. Count how many steps it takes to reach 0 or below, and print the
total subtracted.

```python
n = 100
steps = 0
total_subtraced = 0

while n > 0:
    n -= 7
    steps =+1
    total_subtraced += 7

print(f"Steps taken: {steps}")
print(f"Total subracted: {total_subtraced}")
print(f"Final value of n: {n}")
```

4. The Collatz rule: if n is even, divide by 2. If n is odd, multiply by 3 and add 1. Repeat until n reaches 1. Starting
from n = 27, count how many steps it takes.

```python
n =21
steps = 0

while n != 1:
    if n % 2 == 0:
        n = n // 2
    else:
        n = n*3 +1
    steps += 1

print(f"Steps to reach 1: {steps}")
```

```python
n = 1
hours = 0

while n <= 1000000:
    n = n *2
    hours +=1

print(hours)
```

6. Given nums = [10, 25, 8, 42, 17, 33, 6], sum all elements using a while loop with an index. Then
rewrite it with a for loop and confirm same result.

7. Given students = ["Ali", "Sara", "Rahim", "Mia", "Karim"], search for "Rahim" and print
which position (1-based index) he is at. Stop as soon as found.

```python
students = ["Ali", "Sara", "Rahim", "Mia", "Karim"]
found = False
i = 0
target = "Rahim"

while i < len(students):
    if students[i] == target:
        found = True
        break
    i += 1

if found:
    print(f"Found '{target}' at position {i +1}")
```

8. Given data = [3, -1, 7, -4, 2, -8, 5, 9], remove all negative numbers from the list using a while
loop. Important: when you remove an item, don't increment i.

```python
data = [3, -1, 7, -4, 2, -8, 5, 9]
i = 0

while i < len(data):
    if data[i] < 0:
        data.pop(i)
    else:
        i += 1

print(f"Cleaned: {data}")
```

9. Given scores = [90, 85, 78, 92, 55, 60, 45, 88], keep adding scores one by one. Stop the
moment the running average drops below 70. Print how many scores were included.


```python
scores = [90, 85, 78, 92, 55, 60, 45, 88]
total = 0
count = 0
i = 0

while i < len(scores):
    total += scores[i]
    count += 1
    avg = total / count

    if avg < 70:
        print(f"Average dropped to {round(avg,1)} after {count} scores.")
        break
    i += 1
else:
    print(f"All scores included. Final avg: {round(avg,1)}")
```

10. Given a sorted list nums = [1, 3, 5, 7, 9, 11, 13] and target 14, find two numbers that add up to 14.
Use a left pointer starting at 0 and a right pointer starting at the end.

```python
nums = [1, 3, 5, 7, 9, 11, 13]
target = 14
left = 0
right = len(nums) - 1

while left < right:
    current_sum = nums[left] + nums[right]

    if current_sum == target:
        print(f"Found: {nums[left]} + {nums[right]} = {target}")
        break
    elif current_sum < target:
        left += 1
    else:
        right -= 1
```

12. From stock = {"rice": 50, "oil": 8, "salt": 200, "flour": 0, "sugar": 30}, find the first
item with zero stock and stop immediately.

```python
stock = {"rice": 50, "oil": 8, "salt": 200, "flour": 0, "sugar": 30}
keys = list(stock.keys())
i = 0

while i < len(keys):
    key = keys[i]
    if stock[key] == 0:
        out_of_stock = key
        break
    i += 1
print(out_of_stock)
```

13. You can only afford items up to a total of 500. Given items =
[("pen",15),("notebook",80),("bag",350),("eraser",5),("ruler",30)], add items to a cart
dict one by one until the next item would exceed your budget.


```python
items = [
    ("pen",15),
    ("notebook",80),
    ("bag",350),
    ("eraser",5),
    ("ruler",30)
    ]
cart = {}
total = 0
i = 0

while i < len(items):
    name, price = items[i]
    if total + price > 500:
        break
    cart[name] = price
    total += price
    i += 1

print(cart)
print(total)
```

14. Given queue = {"task1": "send email", "task2": "update db", "task3": "notify user"},
process and remove each task one by one using a while loop until the dict is empty.

```python
queue = {
    "task1": "send email", 
    "task2": "update db", 
    "task3": "notify user"
    }
while len(queue) > 0:
    task_id = list(queue.keys())[0]
    action = queue.pop(task_id)
    print(f"Processing {task_id}: {action}")

print("All tasks done. Queue empty:", queue)
```
