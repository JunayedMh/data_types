1. Given items = [200, 450, 150, 800, 300, 120, 600] and a budget of 1000, add items one by one
until the budget runs out. Stop the moment adding the next item would exceed the budget.

```python
items = [200, 450, 150, 800, 300, 120, 600]
budget = 1000
total = 0
i = 0

while i < len(items):
    if total + items[i] > budget:
        break
    total += items[i]
    print(f"Added {items[i]}, running total: {total}")
    i += 1

print(f"total spent: {total}")
```

14. Given scores before and after coaching: before = [55, 72, 80, 45, 68, 91, 60] and after =
[62, 70, 85, 58, 65, 91, 75]. Count how many students improved, stayed same, or dropped.

```python
before = [55, 72, 80, 45, 68, 91, 60]
after = [62, 70, 85, 58, 65, 91, 75]

improved = same = dropped = 0

for befr, aftr in zip(before,after):
    if aftr > befr:
        improved += 1
    elif aftr < befr:
        dropped += 1
    else:
        same += 1

print(f"Improve: {improved}, Same: {same}, Dropped: {dropped}")
```

16. From data = [12.5, -3.0, 8.2, -1.5, 15.0, 9.8, -0.5, 11.3], build a cleaned list with
negatives removed, then normalise it (Min-Max). Full pipeline: filter → normalise.


**What is Normalisation?**
Normalisation is a data preprocessing step that rescales numerical values into a standard range, usually between **0 and 1**.

**Why do we use it?**
It prevents features with naturally larger numbers (like income) from dominating features with smaller numbers (like age) during machine learning model training, ensuring all data points contribute equally.

**How it works (Min-Max Scaling):**
It finds the minimum value in your data and sets it to `0.0`, finds the maximum value and sets it to `1.0`, and shifts all other values proportionally in between using the formula:

$$x_{\text{scaled}} = \frac{x - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}}$$

```python
data = [12.5, -3.0, 8.2, -1.5, 15.0, 9.8, -0.5, 11.3]
clean_data = [value for value in data if value >= 0]

lowest = clean_data[0]
highest = clean_data[0]

for data in clean_data:
    if data > highest: highest = data
    if data < lowest: lowest = data

normalised = [round((data - lowest) / (highest - lowest), 3) for data in clean_data]

print("Clean data:",clean_data)
print("Normalised:", normalised)
```

17. From the orders list, build a summary dict for completed orders only: {order_id: {"amount": x, "tax":
y, "total": z}} where tax is 8%.

```python
orders = [
("ORD1", 1500, "completed"),
("ORD2", 800, "cancelled"),
("ORD3", 2200, "completed"),
("ORD4", 4500, "cancelled"),
("ORD5", 3100, "completed"),
]
summary = {}

for order_id, amount, status in orders:
    if status == "cancelled":
        continue
    tax = round(amount * 0.08,2)
    total = amount + tax

    summary[order_id] = {"amount": amount, "tax": tax, "total": total}

for o_id, info in summary.items():
    print(f"{o_id} : {info}")
```

18. Given student names, before and after scores, build a dict: {name: {"before": x, "after": y,
"change": z, "result": label}}.

```python
names = ["Ali", "Sara", "Rahim", "Mia", "Karim"]
before = [55, 72, 80, 45, 68]
after = [62, 70, 85, 58, 65]

report = {}

for name,b,a in zip(names,before,after):
    change = a-b
    if change > 0:label = "Improved"
    elif change == 0:label == "Same"
    else:label == "Not Improved"
    report[name] = {"before": b, "after": a, "change": change, "result": label}

for name, info in report.items():
    print(f"{name}: {info}")
```

19. From a word frequency dict, build a list of the top 3 most frequent words. text = "python is great and
python is easy and python is fun and coding is fun".


```python
text = "python is great and python is easy and python is fun and coding is fun"
words = text.split()
freq = {}

for word in words:
    freq[word] = freq.get(word,0) + 1

sorted_words = sorted(freq.items(), key=lambda x:x[1], reverse = True)

print(sorted_words)

top3 = []
i = 0

while i < len(sorted_words) and len(top3)<3:
    top3.append(sorted_words[i])
    i += 1

print("Top 3 words:", top3)

for word, count in top3:
    print(f"{word}: {count}x")
```

20. Given text = "data science with python is great python makes data analysis easy and
fun", build a complete analysis: unique words, word frequencies, top 3 words, and a joined summary string.


```python
text = "data science with python is great python makes data analysis easy and fun"
words = text.split()
print(f"Total words: {len(words)}")
print(f"words: {words}")

unique_words = set(words)
print(f"Total unique words: {len(unique_words)}")
print(f"unique_words: {unique_words}")

freq = {}
for word in words:
    freq[word] = freq.get(word,0)+1

sortedd_words = sorted(freq.items(), key=lambda x:x[1],reverse=True)
print(f"sorted_words: {sortedd_words}")

top3 = []
i = 0
while i < len(sortedd_words) and len(top3) <3:
    top3.append(sortedd_words[i])
    i += 1
for word, count in top3:
    print(f"{word}: {count}x")
print()

summary_details = [f"{word}({count})" for word,count in top3]
summary_string = " | ".join(summary_details)
print(summary_details)
```
