1. Given a list of sentences: lines = ["Python is fun", "I love coding", "Data science is
powerful"]. Count the total number of words across all sentences.


```python
lines = ["Python is fun", "I love coding", "Data science is powerful"]

total_words = 0

for line in lines:
    words = line.split()
    total_words += len(words)

print("Total Words: ",total_words)
```

2. You have two lists: items = ["rice", "oil", "sugar"] and prices = [80, 150, 60]. Use zip() to
calculate the total cost.

*zip()* function in Python like a zipper on a jacket. It takes two or more separate lists (or tuples) and pairs up their corresponding elements together, side-by-side.


```python
items = ["rice", "oil", "sugar"]
prices = [80, 150, 60]
total = 0

for item,price in zip(items,prices):
    total += price
    print(f"{item}: {price}")

print("Total cost: ",total)
```

```python
item_price = {}
total_p = 0
for i in range(len(items)):
    item_price[items[i]] = prices[i]
print(item_price)

for item,price in item_price.items():
    total_p += price

print("Total price: ",total_p)
```

4. Given cart = {"shirt": 500, "pants": 900, "shoes": 1200, "cap": 300} and a 15% discount
on everything, calculate total amount saved.

```python
cart = {"shirt": 500, "pants": 900, "shoes": 1200, "cap": 300}
total = 0

for item,price in cart.items():
    total += price
    discount = total * (15/100)

print("Total saved:",discount)
```
