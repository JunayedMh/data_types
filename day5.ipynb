3. Given scores = (15, 20, 8, 18, 25, 12, 22), calculate the total score after dropping the single
lowest round.

```python
scores = (15, 20, 8, 18, 25, 12, 22)
total = 0
lowest = scores[0]

for score in scores:
    total += score
    if score < lowest:
        lowest = score

final = total - lowest

print(final)
```

4. Given sales = {"phone": [12000, 15000, 9000], "laptop": [45000, 52000], "tablet":
[8000, 9500, 11000, 7500]}, calculate each product's revenue and the overall grand total.

```python
sales = {"phone": [12000, 15000, 9000], "laptop": [45000, 52000], "tablet": [8000, 9500, 11000, 7500]}

grand_total = 0

for product, sale in sales.items():
    product_total = 0
    for amount in sale:
        product_total += amount
    grand_total += product_total
    print(f"{product}: {product_total}")

print(f"Grand total: {grand_total}")
```

5. Subjects have different weights: grades = [85, 72, 90, 78] and weights = [0.3, 0.2, 0.3,
0.2]. Calculate the weighted average.

```python
grades = [85, 72, 90, 78]
weights = [0.3, 0.2, 0.3,0.2]

w_avg = 0

for grade,weight in zip(grades,weights):
    w_avg += grade * weight

print(w_avg)
```

10. Given scores = (55, 82, 91, 67, 78, 95, 60) and a target score of 75, find the score closest to the
target.

```python
scores = (55, 82, 91, 67, 78, 95, 60)
target = 75
closest = scores[0]
second_closest = scores[1]
third_closest = scores[2]

for score in scores:
    if abs(score - target) < abs(closest - target):
        closest = score
    elif abs(score - target) <abs(second_closest - target) and score!= closest: # finding out the 2nd closest
        second_closest = score
    elif abs(score - target) <abs(third_closest- target) and score!= closest and second_closest: # finding out the 3rd closest
        third_closest = score

print(closest)
print(second_closest)
print(third_closest)
```

14. From sales = {"phone": [12000, 15000, 9000], "laptop": [45000, 0, 52000], "tablet":

```python
sales = {"phone": [12000, 15000, 9000], "laptop": [45000, 0, 52000], "tablet": [8000, 9500, 11000]}
always_sold = 0

for product, amounts in sales.items():
    sold_every_period = True
    for amt in amounts:
        if amt == 0:
            sold_every_period = False
    if sold_every_period:
        always_sold += 1
        print(f" ALways sold: {product}")

print(always_sold)
```

15. Given text = "to be or not to be that is the question to ask", count how many distinct
words appear more than once. Use dict.get().

```python
text = "to be or not to be that is the question to ask"
words = text.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) +1

repeated = 0
for word, count in freq.items():
    if count > 1:
        repeated +=1
        print(f"{word}: {count}")


print(freq)
```

17. From students = [("Ali", 80), ("Sara", 45), ("Rahim", 74), ("Mia", 38), ("Karim",

```python
students = [("Ali", 80), ("Sara", 45), ("Rahim", 74), ("Mia", 38), ("Karim", 91)]

passed_names = []

for names, marks in students:
    if marks >= 50:
        passed_names.append(names)

result = ", ".join(passed_names)

print(result)
```

19. From scores = (55, 82, 91, 67, 78, 95, 60, 45, 88, 73), build a dict grouping scores into
bands: {"A": [...], "B": [...], "C": [...], "F": [...]}.

```python
scores = (55, 82, 91, 67, 78, 95, 60, 45, 88, 73)
bands = {"A": [], "B": [], "C": [], "F": []}

for score in scores:
    if score >= 90:
        bands["A"].append(score)
    elif score >=75:
        bands["B"].append(score)
    elif score >= 60:
        bands["C"].append(score)
    else:
        bands["F"].append(score)

for band, values in bands.items():
    print(f"{band}: {values}")

print(bands)
```

20. From hours = {"Mon": 8, "Tue": 7.5, "Wed": 8, "Thu": 6, "Fri": 7, "Sat": 4}, build two
lists — overwork days (above 7h) and light days (7h or below) — then print each as a joined string.

```python
hours = {"Mon": 8, "Tue": 7.5, "Wed": 8, "Thu": 6, "Fri": 7, "Sat": 4}

overwork_days = []
light_days = []

for days, hour in hours.items():
    if hour > 7:
        overwork_days.append(days)
    else:
        light_days.append(days)

print("Overwork days:", ", ".join(overwork_days))
print("Light days:", ", ".join(light_days))
```
