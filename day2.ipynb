```python
# A football team's goals per match are stored as goals = (2, 0, 3, 1, 4, 2, 1). Calculate the total goals scored across all matches.

goals = (2, 0, 3, 1, 4, 2, 1)
total = 0

for goal in goals:
    total += goal

print("Total goal: ", total)
```

```python
# Given expenses = {"rent": 8000, "food": 3500, "transport": 1200, "internet": 500,"misc": 900}, calculate total monthly spending.

expenses = {"rent": 8000, "food": 3500, "transport": 1200, "internet": 500,"misc": 900}
monthly_total = 0

for exp, amt in expenses.items():
    monthly_total += amt

print("Total spendings: ", monthly_total)
```

```python
# Given nums = [1, 2, 3, 4, 5], calculate the sum of squares.

nums = [1, 2, 3, 4, 5]
sum_sq = 0

for num in nums:
    sum_sq += num ** 2

print("total: ", sum_sq)
```

```python
# Sales records are stored as sales = [("Mon", 1500), ("Tue", 2300), ("Wed", 1800), ("Thu",2700), ("Fri", 3100)]. Find the total weekly sales.

sales = [("Mon", 1500), ("Tue", 2300), ("Wed", 1800), ("Thu",2700), ("Fri", 3100)]
sale_dict = dict(sales)
total_wk = 0

for day,sale in sale_dict.items():
    total_wk += sale

print("Total sale: ", total_wk)
```

```python
# Given word = "mississippi", find how many unique letters it contains. Then print the total character count vs unique count.

word = "mississippi"
unique = set()

for ch in word:
    unique.add(ch)
print("Unique: ", unique)
```

```python
# Given matches = [("Match 1", 2), ("Match 2", 5), ("Match 3", 1), ("Match 4", 4)], find which match had the most goals

matches = [("Match 1", 2), ("Match 2", 5), ("Match 3", 1), ("Match 4", 4)]

highest_goal = 0
h_match = ""
for match, goal in matches:
    if goal > highest_goal:
        highest_goal = goal
        h_match = match
print(f"{h_match} has the highest goal: {highest_goal}")
```

```python
# Given fares = {"Monday": 850, "Wednesday": 620, "Friday": 950, "Sunday": 580}, find the cheapest day to travel

fares = {"Monday": 850, "Wednesday": 620, "Friday": 950, "Sunday": 580}
cheapest = 100000000

for day, price in fares.items():
    if price < cheapest:
        cheapest = price

print(f"{day} has the cheapest price: {cheapest}$")
```

```python
# Given words = ("cat", "elephant", "ox", "rhinoceros", "bee"), find the longest word without using max().

words = ("cat", "elephant", "ox", "rhinoceros", "bee")
len_word = 0
longest_word = ""

for word in words:
    if len(word) > len_word:
        longest_word = word
        len_word = len(word)
print(longest_word)
```
