1. Write a function that return how many total times a substring appears accross every line.

```python
def total_occurences(lines, substring):
    total = 0

    for line in lines:
        total += line.count(substring)
    return total

logs = ["ERROR: disk full", "INFO: system ok",
"ERROR: disk read fail", "WARNING: low memory", "ERROR: network down"]

print(total_occurences(logs,"ERROR"))
```

```python
def total_occurences(string, substring):
    total = 0
    string_len = len(string)
    sub_len = len(substring)

    for i in range(string_len - sub_len+1):
        if string[i:i+sub_len] == substring:
            total +=1
    return total

string = input().strip()
substring = input().strip()

print(total_occurences(string,substring))
```

2. Write a function that sums the character length of every username starting with a given letter.

```python
def total_length_startings_with(usernames, letter):
    total = 0

    for name in usernames:
        if name[0] == letter:
            total += len(name)
    return total

usernames = ["alice92", "bob_smith", "amanda_k","charlie99", "aaron_t"]
print(total_length_startings_with(usernames, "a"))
```

3. Write a function that walks a mixed string character by character and sums only the digit characters.

```python
def sum_of_digit(string):
    total = 0
    for s in string:
        if s.isdigit():
            total += int(s)
    return total

print(sum_of_digit("a1b22c333"))
```

4. Write a function that return the combines count of 'G' and 'C' bases across all sequences.

```python
def total_gc_count(seqs):
    total = 0

    for seq in seqs:
        total += seq.count("G") + seq.count("C")
    return total

seqs = ["ATCGGC", "GGATCC", "TTAGCC", "CGCGAT"]
print(total_gc_count(seqs))
```

```python
def total_gc_counts(seqs):
    total = 0

    for seq in seqs:
        for s in seq:
            if s == "G" or s == "C":
                total+=1
    return total

seqs = ["ATCGGC", "GGATCC", "TTAGCC", "CGCGAT"]
print(total_gc_counts(seqs))
```

5. Write a function that takes a list of (filename, size) tuples and sums the sizes of files ending with the given extension.

```python
def total_size_by_extension(files, ext):
    total = 0

    for name, size in files:
        if name.endswith(ext):
            total += size
    return total

files = [("report.pdf", 1200), ("data.csv", 340), ("notes.txt", 50), ("archive.zip", 900), ("summary.pdf", 800)]

print(total_size_by_extension(files, ".pdf"))
```

```python
def total_amount_by_size(files,ext):
    total = 0

    for name, size in files:
        if name[-4:] == ext:
            total += size
    return total

files = [("report.pdf", 1200), ("data.csv", 340), ("notes.txt", 50), ("archive.zip", 900), ("summary.pdf", 800)]

print(total_amount_by_size(files, ".pdf"))
```

6. Write a function that returns the longest username using the tracket pattern.

```python
def longest_username(usernames):
    longest = usernames[0]

    for name in usernames:
        if len(name) > len(longest):
            longest = name
    return longest

print(longest_username(["alice92","bob_smith","amanda_k","charlie99","aaron_t"]))
```

9. Write is_palindrome_two_pointer(s). Instead of reversing the string, place a pointer at each END and
walk them toward the middle, comparing as you go

```python
def is_palindrome_two_pointers(s):
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -=1
    return True

print(is_palindrome_two_pointers("racecar"))
print(is_palindrome_two_pointers("Hello"))
```

10. Write a function that finds, for each word, its most frequent character count (using .count() per unique character), then tracks the word with the highest.

```python
def word_with_most_repeated_char(words):
    the_word = words[0]
    appeard = 0

    for word in words:
        max_char_count = 0
        for char in word:
            ch = word.count(char)
            if ch > max_char_count:
                max_char_count = ch
                if max_char_count > appeard:
                    appeard = max_char_count
                    the_word = word
    return the_word,appeard

print(word_with_most_repeated_char(["hello","banana","mississippi","book"]))
```

11. Write a function that return the sequence whose length is closest to the target, using distance comparison instead of raw size.

```python
def closest_length(seqs, target):
    closest = seqs[0]
    closest_seq = []
    for seq in seqs:
        if abs(len(seq)-target) < abs(len(closest)-target):
            closest = seq
    
    min_diff = abs(len(closest) - target)
    for seq in seqs:
        if abs(len(seq)-target) == min_diff:
            closest_seq.append(seq)
    return closest_seq

seqs = ["ATCG", "GATTACA", "CCGT", "ATCGGGT"]
print(closest_length(seqs, 6))
```

12. Write a function that counts how many plaindrome words in a list. (use slicing or two pointer)

```python
def is_palindrome(word):
    if word[0:] == word[::-1]:
        return word


def count_palindrome(words):
    total = 0
    for word in words:
        if is_palindrome(word):
            total += 1
    return total

print(count_palindrome(["level","hello","racecar","world","noon"]))
```

13. Write a function that counts how many strings contain substring MORE than n times.

```python
def count_strings_with_repeats(strings, substring, n):
    total = 0
    for string in strings:
        if string.count(substring) > n:
            total += 1
    return total

logs = ["error error warning", "error info", "error error error", "info info"]
print(count_strings_with_repeats(logs, "error", 1))
```
