4. The Collatz rule: if n is even, divide by 2. If n is odd, multiply by 3 and add 1. Repeat until n reaches 1. Starting
from n = 27, count how many steps it takes.

```python
n =21
steps = 0

while n != 1:
    if n % 2 == 0:
        n = n // 2
    else:
        n = n*3 +1
    steps += 1

print(f"Steps to reach 1: {steps}")
```

9. Given scores = [90, 85, 78, 92, 55, 60, 45, 88], keep adding scores one by one. Stop the
moment the running average drops below 70. Print how many scores were included.


```python
scores = [90, 85, 78, 92, 55, 60, 45, 88]
total = 0
count = 0
i = 0

while i < len(scores):
    total += scores[i]
    count += 1
    avg = total / count

    if avg < 70:
        print(f"Average dropped to {round(avg,1)} after {count} scores.")
        break
    i += 1
else:
    print(f"All scores included. Final avg: {round(avg,1)}")
```

10. Given a sorted list nums = [1, 3, 5, 7, 9, 11, 13] and target 14, find two numbers that add up to 14.
Use a left pointer starting at 0 and a right pointer starting at the end.

```python
nums = [1, 3, 5, 7, 9, 11, 13]
target = 14
left = 0
right = len(nums) - 1

while left < right:
    current_sum = nums[left] + nums[right]

    if current_sum == target:
        print(f"Found: {nums[left]} + {nums[right]} = {target}")
        break
    elif current_sum < target:
        left += 1
    else:
        right -= 1
```

13. You can only afford items up to a total of 500. Given items =
[("pen",15),("notebook",80),("bag",350),("eraser",5),("ruler",30)], add items to a cart
dict one by one until the next item would exceed your budget.


```python
items = [
    ("pen",15),
    ("notebook",80),
    ("bag",350),
    ("eraser",5),
    ("ruler",30)
    ]
cart = {}
total = 0
i = 0

while i < len(items):
    name, price = items[i]
    if total + price > 500:
        break
    cart[name] = price
    total += price
    i += 1

print(cart)
print(total)
```

15. Given prices = {"apple": 50, "banana": 20, "mango": 80, "grape": 150, "kiwi": 200},
apply a 10% discount to each item one by one using a while loop. Stop applying discounts after the running
total savings exceeds 30

```python
prices = {"apple": 50, "banana": 20, "mango": 80, "grape": 150, "kiwi": 200}
keys = list(prices.keys())
savings = 0
i = 0

while i <len(keys) and savings <= 30:
    key = keys[i]
    discount = prices[key] * 0.10
    prices[key] = round(prices[key] - discount, 2)
    savings += discount 
    print(f"Discounted {key}; saved {discount:.1f}, total saved: {savings:.1f}")
    i += 1

print("Updated prices:",prices)
```

16. You have stock = {"phone": 3, "laptop": 1, "tablet": 2} and incoming orders = ["phone","tablet","laptop","phone","tablet","phone","laptop"]. Process each order,
deduct from stock. Skip if out of stock. Stop when all stock is zero.


```python
stock = {"phone": 3, "laptop": 1, "tablet": 2}
orders = ["phone","tablet","laptop","phone","tablet","phone","laptop"]

i = 0

while i < len(orders) and any(qty > 0 for qty in stock.values()):
    item = orders[i]

    if stock.get(item,0)> 0:
        stock[item] -=1
        print(f"Fulfilled: {item} | Remaining: {stock[item]}")
    else:
        print(f"Out of stock: {item}")
    i += 1

print("Final stock:", stock)
```

17. Given updates = [("Ali",10),("Sara",15),("Ali",8),("Rahim",30),("Sara",5),("Ali",20)], process score
updates one by one. Stop the moment the leader changes from whoever was leading after the first update.


```python
updates = [
    ("Ali",10),
    ("Sara",15),
    ("Ali",8),
    ("Rahim",30),
    ("Sara",5),
    ("Ali",20)
    ]
scores = {}
leader = None
i = 0

while i < len(updates):
    name,pts = updates[i]
    scores[name] = scores.get(name,0)+ pts 
    current_leader = max(scores, key=lambda k:scores[k])
    if leader is None:
        leader = current_leader
        print(f"Initital leader: {leader}")
    elif current_leader !=leader:
        print(f"Leader changed to {current_leader} after update {i+1}!")
        leader = current_leader
        break
    i += 1

print("Final scores so far:",scores)
```

18. Given accounts = {"ACC1": 5000, "ACC2": 1200, "ACC3": 8000} and withdrawals =
[("ACC1",2000),("ACC2",1500),("ACC3",3000),("ACC1",4000),("ACC3",1000)], process each
withdrawal. Reject if insufficient funds.


```python
accounts = {"ACC1": 5000, "ACC2": 1200, "ACC3": 8000}
withdrawals = [("ACC1",2000),("ACC2",1500),("ACC3",3000),("ACC1",4000),("ACC3",1000)]
i = 0

while i < len(withdrawals):
    acc,amount = withdrawals[i]
    if accounts.get(acc,0) >= amount:
        accounts[acc] -= amount
        print(f"OK   {acc}: Withdrew {amount}, balance: {accounts[acc]}")
    else:
        print(f"REJECT {acc}: amount exceeded current balance, the {acc} has {accounts.get(acc,0)}")
    i += 1

print(accounts)
```

19. Given text = "the cat and the dog and the bird and the fish", process words one by one
building a frequency dict. Stop the moment any word appears 3 times. Print which word triggered the stop

```python
text = "the cat and the dog and the bird and the fish"
words = text.split()
freq = {}
stop_word = None
i = 0

while i < len(words):
    word = words[i]
    freq[word] = freq.get(word, 0) +1
    if freq[word] >=3:
        stop_word = word
        break
    i +=1

print(f"Stopped at word #{i+1}: '{stop_word}' appeared {freq[stop_word]} times")
print(f"Freq dict so far: {freq}")
```
