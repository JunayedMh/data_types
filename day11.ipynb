1. Write a function that pushes every value onto a stack, then drain it with.pop(), summing as it goes. 

```python
def sum_stack(numbers):
    stack = []
    for num in numbers:
        stack.append(num)

    total = 0
    while stack:
        total += stack.pop()
    return total

print(sum_stack([12,8,15,20,5]))
```

2. Write a function that pushes everything onto a stack, then pops exactly n times and sumy only those.

```python
def sum_stack_n(numbers, n):
    stack = []
    for number in numbers:
        stack.append(number)
    
    total = 0
    for _ in range(n):
        total += stack.pop()
    return total

print(sum_stack_n([68,70,65,72,74,69],3))
```

3. A scorekeeping game processes operations one at a time: a number pushes that score, "C" cancels the last score, "D" doubles the last score and adds it as new, "+" adds the last two scores together as a new entry. Write a function that return the sum of scores remaining after processing every operation.

```python
def scoreboard(scores):
    board = []
    for score in scores:
        if score == "C":
            board.pop()
        elif score == "D":
            board.append(board[-1] * 2)
        elif score == "+":
            board.append(board[-1] + board[-2])
        else:
            board.append(int(score))
    return sum(board)

print(scoreboard(["5", "2", "C", "D", "+"]))
```

4. Write a funtion that pushes weights onto a stack one at a time, but if adding the next weight would exceed capactity, pop from the top until it fits. eg:  [4, 6, 5, 3, 8, 2], capacity=10 → 10

```python
def total_weights(weights, capacity):
    stack = []
    for weight in weights:
        while stack and (sum(stack) + weight> capacity):
            stack.pop()
        stack.append(weight)
    return sum(stack)

print(total_weights([4,8,5,3,8,2],10))
```

5. Write a function. Push each value, but if the new value cancels out the current top(they sum to zero), pop instead of pushing. Return the sum of what's left.

```python
def sum_after_cancel_pairs(values):
    stack = []
    for value in values:
        if stack and (stack[-1]+value) == 0:
            stack.pop()
        else:
            stack.append(value)
    return sum(stack)
print(sum_after_cancel_pairs([3, -3, 5, 7, -7, 2]))
```

6. Write a function that pushes each value onto a list, and at every step records the highest values seeen SO FAR. e.g:  [4, 9, 2, 15, 7] → [4, 9, 9, 15, 15]


```python
def keep_the_highest(values):
    highest_list = []
    highest = 0
    for value in values:
        if value > highest:
            highest = value
        highest_list.append(highest)
    return highest_list

print(keep_the_highest([4,9,9,3,2,15,15]))
```
