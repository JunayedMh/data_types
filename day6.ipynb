1. Raw sales data has currency symbols: raw = ["$120", "$350", "$89", "$450", "$75"]. Strip the $
sign, convert to numbers and calculate the total.


```python
raw = ["$120", "$350", "$89", "$450", "$75"]
total = 0

for item_price in raw:
    cleaned = item_price.replace("$", "")
    total += float(cleaned)

print(total)
```

3. Each round has points and a penalty flag: rounds = [("R1", 50, False), ("R2", 80, True),
("R3", 60, False), ("R4", 90, True), ("R5", 40, False)]. Penalty rounds lose 20 points.
Calculate the true total.

```python
rounds = [("R1", 50, False), ("R2", 80, True),
("R3", 60, False), ("R4", 90, True), ("R5", 40, False)]

total = 0

for name, points, penalty in rounds:
    if penalty:
        total += points - 20
    else:
        total += points

print("Total:",total)
```

6. Given raw = ["$120", "$350", "$89", "$450", "$75"], find the highest price after stripping the $symbol.

18. From results = {"Ali": [78, 82, 55], "Sara": [90, 88, 76], "Rahim": [45, 70, 60],
"Mia": [80, 75, 91]}, build a dict mapping each student to "All Pass", "All Fail", or "Mixed".

```python
results = {"Ali": [78, 82, 55], "Sara": [90, 88, 76], "Rahim": [45, 70, 60], "Mia": [80, 75, 91]}

report = {}
for student, marks in results.items():
    if all(mark>= 50 for mark in marks):
        report[student] = "All Pass"
    elif all(mark <50 for mark in marks):
        report[student] = "All Fail"
    else:
        report[student] = "Mixed"

print(report)
```

20. From orders = {"O1": (500, "complete"), "O2": (300, "pending"), "O3": (800,
"complete"), "O4": (200, "cancelled"), "O5": (650, "complete")}, build a summary dict:

```python
orders = {"O1": (500, "complete"), 
          "O2": (300, "pending"), 
          "O3": (800,"complete"), 
          "O4": (200, "cancelled"), 
          "O5": (650,"complete")
          }

revenue = 0
status_count = {}
complete_ids = []

for order_id, (amount,status) in orders.items():
    if status == "complete":
        revenue += amount
        complete_ids.append(order_id)
    
    status_count[status] = status_count.get(status,0) +1

summary = {
    "total_revenue": revenue,
    "Status_counts": status_count,
    "Completed_ids": complete_ids
}


for key,val in summary.items():
    print(f"{key}: {val}")
```
