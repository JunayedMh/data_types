1. Write a function that pushes every value onto a stack, then drain it with.pop(), summing as it goes. 

```python
def sum_stack(numbers):
    stack = []
    for num in numbers:
        stack.append(num)

    total = 0
    while stack:
        total += stack.pop()
    return total

print(sum_stack([12,8,15,20,5]))
```

2. Write a function that pushes everything onto a stack, then pops exactly n times and sumy only those.

```python
def sum_stack_n(numbers, n):
    stack = []
    for number in numbers:
        stack.append(number)
    
    total = 0
    for _ in range(n):
        total += stack.pop()
    return total

print(sum_stack_n([68,70,65,72,74,69],3))
```

3. A scorekeeping game processes operations one at a time: a number pushes that score, "C" cancels the last score, "D" doubles the last score and adds it as new, "+" adds the last two scores together as a new entry. Write a function that return the sum of scores remaining after processing every operation.

```python
def scoreboard(scores):
    board = []
    for score in scores:
        if score == "C":
            board.pop()
        elif score == "D":
            board.append(board[-1] * 2)
        elif score == "+":
            board.append(board[-1] + board[-2])
        else:
            board.append(int(score))
    return sum(board)

print(scoreboard(["5", "2", "C", "D", "+"]))
```
