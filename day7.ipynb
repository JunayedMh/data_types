1. Given items = [200, 450, 150, 800, 300, 120, 600] and a budget of 1000, add items one by one
until the budget runs out. Stop the moment adding the next item would exceed the budget.

```python
items = [200, 450, 150, 800, 300, 120, 600]
budget = 1000
total = 0
i = 0

while i < len(items):
    if total + items[i] > budget:
        break
    total += items[i]
    print(f"Added {items[i]}, running total: {total}")
    i += 1

print(f"total spent: {total}")
```

2. A sensor log has some corrupted negative readings: data = [12.5, -3.0, 8.2, -1.5, 15.0, 9.8,
-0.5, 11.3]. Sum only the valid positive readings using continue to skip negatives.


```python
data = [12.5, -3.0, 8.2, -1.5, 15.0, 9.8, -0.5, 11.3]
total = 0

for readings in data:
    if readings < 0:
        continue
    total += readings

print(f"Valid total: {total}")
```

3. Starting from 10, count down to 1 and sum all the numbers along the way. Use a while loop with a
decrementing counter.


```python
n = 10
total = 0

while n>=1:
    total += n
    n -= 1
print(f"sum of 10 to 1: {total}")
```

4. Order records have a status field: orders = [("ORD1",1500,"completed"),("ORD2",800,"cancell
ed"),("ORD3",2200,"completed"),("ORD4",450,"cancelled"),("ORD5",3100,"completed")]
. Sum only completed orders using continue

```python
orders = [
    ("ORD1",1500,"completed"),
    ("ORD2",800,"cancelled"),
    ("ORD3",2200,"completed"),
    ("ORD4",450,"cancelled"),
    ("ORD5",3100,"completed")
    ]
total = 0

for order_id, amount, status in orders:
    if status == "cancelled":
        continue
    total += amount

print(f"Total compelte sale: {total}")
```

5. Given donations = [500, 1200, 800, 2500, 300, 1800, 900] and a fundraising target of 5000, add
donations in order and stop the moment the target is reached. Print how many donations it took.

6. From orders = [450, 800, 1500, 2200, 3100, 900, 1800], find the first order amount that exceeds 1000. Stop as soon as you find it, no need to check the rest.

7. From data = [12.5, -3.0, 8.2, -1.5, 15.0, 9.8, -0.5, 11.3], find the highest valid reading,
skipping negatives with continue.

```python
data = [12.5, -3.0, 8.2, -1.5, 15.0, 9.8, -0.5, 11.3]
highest = None

for readings in data:
    if readings < 0:
        continue
    if highest is None or readings > highest:
        highest = readings

print(highest)
```
