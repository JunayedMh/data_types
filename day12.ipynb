1. Write a function that returns the sum of first elements.

```python
def window_sum(nums,k):
    return sum(nums[:k])

print(window_sum([2,5,1,8,3,9],3))
```

```python
def index_window_sum(nums,k,start):
    return sum(nums[start:start+k])

print(index_window_sum([2,5,1,8,3,9],3,2))
```

2. Write a function that returns a list of every k-length window's sum, but computes the first sum once  and then slides: subtract the value leaving on the left, add the value entering on the right. 

Example: [2, 5, 1, 8, 3, 9], k=3 → [8, 14, 12, 20]

```python
def all_window_sums(nums,k):
    window_sum = sum(nums[:k])
    sums = [window_sum]
    for i in range(k,len(nums)):
        window_sum += nums[i] - nums[i-k]
        sums.append(window_sum)
    return sums

print(all_window_sums([2, 5, 1, 8, 3, 9],3))
```
