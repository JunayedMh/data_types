```python
# A daily meal plan has calories per meal: meals = [450, 320, 580, 210, 390]. Calculate total calories consumed.

meals = [450, 320, 580, 210, 390]
total_cl = 0

for calore in meals:
    total_cl += calore

print("Total Calories: ",total_cl)
```

```python
# A quiz player's scores per round are rounds = (15, 20, 8, 18, 25, 12). Find their total score and average score per round.

rounds = (15, 20, 8, 18, 25, 12)
total = 0

for score in rounds:
    total += score

avg = total/ len(rounds)

print("Total: ",total)
print("Average: ",round(avg,2))
```
