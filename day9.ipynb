1. Write a function that takes a list of prices and returns the sum. 

```python
def total_price(prices):
    if not prices:
        return None
    total = 0
    for price in prices:
        total += price
    return total

print(total_price([250,454,324]))
print(total_price([]))
```

2. Write a function that returns the total after subtracting a discount percentage. If no discount is given, it should behave like a plain total.

```python
def total_with_discount(prices, discount=0):
    total = 0

    for price in prices:
        total += price
    return total - (total*discount/100)

print(total_with_discount([123,454], discount=10))
print(total_with_discount([843,322]))
```

3. Write a helper function that returns square of a number. Then write sum_of_squares(numbers) that calls square() inside its loop instead of writing x*x directly.

```python
def square(num):
    return num*num

def sum_of_squares(numbers):
    total = 0
    for number in numbers:
        total += square(number)
    return total

print(sum_of_squares([1,2,3,4]))
```

6. Write a function to find out the max.

```python
def find_max(nums):
    max_num = nums[0]

    for num in nums:
        if num > max_num:
            max_num=num
    return max_num

print(find_max([5,8,9,6,7,3]))
```

7. Write a function to find out the min number with index.

```python
def find_min(numbers):
    min_num = numbers[0]

    for i in range(len(numbers)):
        if numbers[i] < min_num:
            min_num = numbers[i]
            index_num = i
    return (min_num,index_num)

print(find_min([4,3,6,2,6,2,5,22,3]))
```

8. Write a helper function that return average. Then write another function that calles it and counts how many values are above that average.

```python
def calc_avg(numbers):
    return sum(numbers)/len(numbers)

def count_above_avg(numbers):
    avg = calc_avg(numbers)
    count = 0

    for num in numbers:
        if num > avg:
            count+=1
    return count

print(count_above_avg([32,34,21,5,43]))
```

10. Write a function that returns True if two number in the list add up to the target. {Don't use nested loop}

```python
def catch_target(numbers,target):
    for i in range(len(numbers)-1):
        if numbers[i] + numbers[i+1] == target:
            return True
    return False

print(catch_target([1,2,7,3],9))
print(catch_target([1,7,5,2],9))

# The logic breaks if the int values are not situated one after another to its target, eg: print(catch_target([1,7,5,2],9))
```

```python
def catch_target(numbers,target):
    seen_numbers = {}

    for num in numbers:
        complement = target - num
        if complement in seen_numbers:
            return True
        seen_numbers[num] = True
    return False


print(catch_target([1,7,5,2],9))   
print(catch_target([6,9,3,4,2,8,1,9,74,6,4,3,9,3,4,3,9,1,8,3],10))
```

11. Write a helper function that returns Vowels. Then write another function to count vowels using the helper function inside the loop.

```python
def is_vowel(char):
    return char in ('a','e','i','o','u')

def count_vowel(word):
    count = 0

    for char in word.lower():
        if is_vowel(char):
            count +=1
    return count

print(count_vowel('Python'))
```

12. Write a fucntion tha builds a new list containing only numbers above the threshold, doubled.

```python
def above_threshold(numbers,threshold):
    new_list = []

    for num in numbers:
        if num > threshold:
            new_list.append(num*2)
    return new_list

print(above_threshold([3,4,5,2,33,5,6,7],5))
```

14. 10. Write a function that returns a tuple if two number in the list add up to the target. {Don't use nested loop}

```python
def two_sum_indices(numbers,target):
    seen = {}

    for i,num in enumerate(numbers):
        complement = target - num

        if complement in seen:
            return (seen[complement],i)
        seen[num]= i
    return None

print(two_sum_indices([2,4,5,4],9))
```

```python
def three_digit_sum(numbers,target):
    for i in range(len(numbers)):
        seen = set()
        
        for j in range(i+1,len(numbers)):
            needed_third = target - (numbers[i] + numbers[j])
            
            if needed_third in seen:
                return True
    return False

print(three_digit_sum([2,3,6,4,8],9))
        
```

15. Write helper function that return a dictionary with keys and values from a list. THen write another function to return which key appears the most.

```python
def the_dict(keys):
    freq = {}
    for key in keys:
        freq[key] = freq.get(key,0)+1
    return freq

def find_most_frequnet(keys):
    freq = the_dict(keys)
    print(freq)
    count = 0
    most_Freq = None

    for key,value in freq.items():
        if value > count:
            most_Freq = key
            count = value
    return most_Freq

print(find_most_frequnet(['sun','moon','sun','star','sun']))
```
