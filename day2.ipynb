```python
# Given fares = {"Monday": 850, "Wednesday": 620, "Friday": 950, "Sunday": 580}, find the cheapest day to travel

fares = {"Monday": 850, "Wednesday": 620, "Friday": 950, "Sunday": 580}
cheapest = 100000000

for day, price in fares.items():
    if price < cheapest:
        cheapest = price

print(f"{day} has the cheapest price: {cheapest}$")
```

```python
# Given words = ("cat", "elephant", "ox", "rhinoceros", "bee"), find the longest word without using max().

words = ("cat", "elephant", "ox", "rhinoceros", "bee")
len_word = 0
longest_word = ""

for word in words:
    if len(word) > len_word:
        longest_word = word
        len_word = len(word)
print(longest_word)
```

```python
# Given scores = {"Ali": [80, 75, 90], "Sara": [95, 88, 92], "Rahim": [70, 65, 78]},find who has the highest average score

scores = {"Ali": [80, 75, 90], "Sara": [95, 88, 92], "Rahim": [70, 65, 78]}
top_student = None
top_avg = 0

for student, score in scores.items():
    total = 0
    for mark in score:
        total += mark
    avg = total / len(scores)
    if avg > top_avg:
        top_avg = avg
        top_student = student

print(f"{top_student} has the highest average : {round(top_avg, 2)}")
```

```python
# Given budget = {"rent": 8000, "food": 3500, "transport": 1200, "internet": 500,"misc": 900}, count how many expense categories are above 1000.

budget = {"rent": 8000, "food": 3500, "transport": 1200, "internet": 500,"misc": 900}
above_1k = 0
items = []
for expense,amt in budget.items():
    if amt > 1000:
        items.append(expense)
        above_1k += 1
print("Items are: ", items)
print("Total items above 1000: ", above_1k)
```

```python
# From results = [("Ali", 80), ("Sara", 45), ("Rahim", 74), ("Mia", 38), ("Karim",91)], build a dict where each name maps to "Pass" or "Fail" (pass = 50+).

results = [("Ali", 80), ("Sara", 45), ("Rahim", 74), ("Mia", 38), ("Karim",91)]
result = {}

for name,mark in results:
    if mark >= 50:
        result[name] = "Pass"
    else:
        result[name] = "Fail"

print(result)
```

```python
# Given stock = {"rice": 5, "oil": 23, "salt": 2, "sugar": 18, "flour": 3}, split into two dicts: one for low stock (below 10), one for sufficient stock

stock = {"rice": 5, "oil": 23, "salt": 2, "sugar": 18, "flour": 3}
low_stock = {}
well_stock = {}

for product,amnt in stock.items():
    if amnt > 10:
        well_stock[product] = amnt
    else:
        low_stock[product] = amnt

print(low_stock)
print(well_stock)
```
